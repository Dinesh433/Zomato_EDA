```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
##To check data types
final_df.dtypes
```

```python
final_df.Country.value_counts()
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
##showing top-3 countries by creating a pie chart 
plt.pie(country_values[:3],
    explode=None,
    labels=country_names[:3],
    colors=None,
    autopct="%.1f%%")
```

```python
#observation:Zomato maximum records or transaction are from India followed by USA and UK
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
##groupby along with renaming 0 to rating counts column
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

```python
ratings
```

# #Conclusions/observations from ratings dataset:
1-When ratings are between 4.5 to 4.9----->Excellent
2-When ratings are between 4.0 to 4.4----->very good
3-When ratings are between 3.5 to 3.9----->good
4-When ratings are between 2.5 to 3.4----->average
5-When ratings are between 1.8 to 2.4----->poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Counts',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Counts',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

# Observation:
   1.Not rated count is very high.
   2-Maximum number of ratings are between 2.5 to 3.4.

```python
##Count plot
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
